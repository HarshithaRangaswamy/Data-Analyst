```python
import json
import pandas as pd
import numpy as np
```

```python
#reading json compressed files
receipts_df = pd.read_json('receipts.json.gz', compression='infer', lines =True) # read line by line
users_df = pd.read_json('users.json.gz', compression='infer', lines =True)
brands_df = pd.read_json('brands.json.gz', compression='infer', lines =True)
```

```python
receipts_df.head()
```

```python
#getting to know the datatype of each column
receipts_df.info()
```

```python
#transforming each element of a list to a row replicating index values
receipts_df = receipts_df.explode('rewardsReceiptItemList')
receipts_df.reset_index(inplace=True)
```

```python
receipts_df
```

```python
#as json_normalise does not work with NAN values filling na's with {}
receipts_df['rewardsReceiptItemList'] = receipts_df['rewardsReceiptItemList'].fillna({i: {} for i in receipts_df.index})
```

```python
receipts_df
```

```python
# rewardsReceiptItemList is dict column so converting keys to column headers and values to rows
receiptItemList_df = pd.json_normalize(receipts_df['rewardsReceiptItemList'], record_prefix = 'rewardsReceiptItemList_', errors = 'ignore').add_prefix('rewardsReceiptItemList_')
```

```python
receiptItemList_df.info()
```

```python
#joining the two dataframes 
norm_receipts_df = receipts_df.join(receiptItemList_df, how = 'outer')
```

```python
#dropping the 'rewardsReceiptItemList' coulmn after extracting the data
norm_receipts_df.drop(columns = ['rewardsReceiptItemList'],axis = 1, inplace = True)
```

```python
#The final joined dataset
norm_receipts_df.head()
```

```python
#converting Date fields which are in dictionary format to DateTime format
norm_receipts_df['createDate'][0]
```

```python
#handling NaN values
norm_receipts_df['finishedDate'] = norm_receipts_df['finishedDate'].fillna({i: {} for i in receipts_df.index})
norm_receipts_df['pointsAwardedDate'] = norm_receipts_df['pointsAwardedDate'].fillna({i: {} for i in receipts_df.index})
norm_receipts_df['purchaseDate'] = norm_receipts_df['purchaseDate'].fillna({i: {} for i in receipts_df.index})
```

```python
#extracting data from dict and then converting into DateTime
norm_receipts_df['createDate'] = pd.to_datetime(pd.json_normalize(norm_receipts_df['createDate'])['$date'],unit = 'ms').astype('datetime64[s]')
norm_receipts_df['dateScanned'] = pd.to_datetime(pd.json_normalize(norm_receipts_df['dateScanned'])['$date'],unit = 'ms').astype('datetime64[s]')
norm_receipts_df['finishedDate'] = pd.to_datetime(pd.json_normalize(norm_receipts_df['finishedDate'])['$date'],errors='coerce',unit = 'ms').astype('datetime64[s]')
norm_receipts_df['modifyDate'] = pd.to_datetime(pd.json_normalize(norm_receipts_df['modifyDate'])['$date'],unit = 'ms').astype('datetime64[s]')
norm_receipts_df['pointsAwardedDate'] = pd.to_datetime(pd.json_normalize(norm_receipts_df['pointsAwardedDate'])['$date'],unit = 'ms').astype('datetime64[s]')
norm_receipts_df['purchaseDate'] = pd.to_datetime(pd.json_normalize(norm_receipts_df['purchaseDate'])['$date'],unit = 'ms').astype('datetime64[s]')
```

```python
norm_receipts_df.head()
```

```python
#extracting data from dictionary 
norm_receipts_df['_id'] = pd.json_normalize(norm_receipts_df['_id'])
```

```python
norm_receipts_df.head()
```

```python
users_df.head()
```

```python
#handling NaN values
users_df['lastLogin'] = users_df['lastLogin'].fillna({i: {} for i in users_df.index})
```

```python
#extracting data from dict and then converting into DateTime
users_df['_id'] = pd.json_normalize(users_df['_id'])
users_df['createdDate'] = pd.to_datetime(pd.json_normalize(users_df['createdDate'])['$date'],unit = 'ms').astype('datetime64[s]')
users_df['lastLogin'] = pd.to_datetime(pd.json_normalize(users_df['lastLogin'])['$date'],unit = 'ms').astype('datetime64[s]')
```

```python
#users data after converting the date fields
users_df.head()
```

```python
brands_df.head()
```

```python
# cpg is dict column so converting keys to column headers and values to rows
cpg_df = pd.json_normalize(brands_df['cpg']).add_prefix('cpg.')
```

```python
#joining the two dataframes 
norm_brands_df = brands_df.join(cpg_df, how = 'outer')
```

```python
#extracting data from dictionary 
norm_brands_df['_id'] = pd.json_normalize(norm_brands_df['_id'])
```

```python
#dropping the 'cpg' column after extracting the data
norm_brands_df.drop(columns = 'cpg', axis = 1, inplace = True)
```

```python
#Final dataset 
norm_brands_df.head()
```

```python
#performing SQL querying using SQLAlchemy
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
```

```python
#dropping duplicates with same _id and writing users dataframe to database
users_df.drop_duplicates(subset=['_id']).to_sql('users_data',con=engine)
```

```python
#writing normalised receipts data to database
norm_receipts_df.to_sql('receipts_data',con =engine)
```

```python
#writing normalised brands data to database
norm_brands_df.drop_duplicates(subset=['brandCode']).to_sql('brands_data',con=engine)
```

```python
#Top 5 brands from the given dataset
engine.execute("select r.rewardsReceiptItemList_brandCode, count(*) from receipts_data r where r.rewardsReceiptItemList_brandCode != 'None' and r.purchaseDate > (select max(purchaseDate) from receipts_data)-30 group by r.rewardsReceiptItemList_brandCode order by count(*) desc LIMIT 5").fetchall()
```

```python
#Top 5 brands for the most recent month
engine.execute("select r.rewardsReceiptItemList_brandCode, count(*) from receipts_data r where r.rewardsReceiptItemList_brandCode != 'None' and r.purchaseDate > (SELECT strftime('%Y-%m-%d %H:%M:%S', (select MAX(purchaseDate) from receipts_data),'-1 month')) group by r.rewardsReceiptItemList_brandCode order by count(*) desc LIMIT 5").fetchall()
```
